```python
import math
import seaborn as sns
from sklearn.cluster import KMeans
import warnings
from pylab import rcParams
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
sns.set_color_codes
sns.set_context('poster')
warnings.filterwarnings("ignore")
rcParams['figure.figsize'] = 10,5
```

```python
def load_dataset(name):
    return np.loadtxt(name)
```

```python
#dataset = load_dataset('Stat.txt')
#dataset = load_dataset('ReintroducedMozilla.txt')
dataset = load_dataset('ReintroducedEclipse.txt')
plt.scatter(dataset[:,0], dataset[:,1])
plt.show()
```

```python
dist_points_from_cluster_centre = []
K = range(1,10)
for no_of_clusters in K:
    k_model = KMeans(n_clusters=no_of_clusters)
    k_model.fit(dataset)
    dist_points_from_cluster_centre.append(k_model.inertia_)
```

```python
dist_points_from_cluster_centre
```

```python
plt.plot(K, dist_points_from_cluster_centre)
```

```python
plt.plot(K, dist_points_from_cluster_centre)
plt.plot([K[0],K[8]],[dist_points_from_cluster_centre[0],
        dist_points_from_cluster_centre[8]],'ro-')
plt.show()
```
